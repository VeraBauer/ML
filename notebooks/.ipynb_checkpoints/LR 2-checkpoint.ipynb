### Масштабирование (изменение диапазона) числовых признаков 

```python
#Масштабирование - процесс повышения пригодности функций для обучения путем изменения масштаба

#Это необходимый шаг, потому что в большинстве случаев числовые характеристики не имеют определенного диапазона и отличаются друг от друга. 
#Это сильно искажает результаты таких алгоритмов, как метод k-ближайших соседей, которые учитывают расстояния между измерениями.
#А масштабирование позволяет этого избежать. 
```

**Нормализация**: все значения будут находиться в диапазоне от 0 до 1

Для того, чтобы сделать нормализацию данных нужно посчитать в каждом признаке его минимум (min) и максимум (max), а затем сделать следующее вычисление:

$$x = \frac{x - min}{max - min}$$

После такого преобразования $min = 0, max = 1$.

**Стандартизация**: масштабирует значения с учетом стандартного отклонения.

Для того, чтобы сделать стандартизацию данных нужно посчитать в каждом признаке его среднее значение (`mean`) и стандартное отклонение (`std`), а затем сделать следующее вычисление:

$$x = \frac{x - mean}{std}$$

После такого преобразования $mean = 0, std = 1$.

### Метрики качества

$P$ – число истинных результатов, $P = TP + FN$

$N$ – число ложных результатов, $N = TN + FP.$ 

* tp - количество истинных положительных результатов
* fn - количество ложных отрицательных результатов
* tn - количество истинных отрицательных результатов
* fp - количество ложных положительных результатов

```python
#accuracy = (количество верно классифицированных объектов) / (общее количество объектов)
print("Accuracy :", accuracy_score(y_test, knn_y_pred)) #Оценка точности классификации

#The best value is 1 and the worst value is 0.
#precision = tp / (tp + fp)
#recall = tp / (tp + fn)
#F1 - среднее гармоническое между precision и recall
#F1 = 2 * (precision * recall) / (precision + recall)
print(classification_report(y_test, knn_y_pred))
```

ROC — Receiver Operating Characteristic (рабочая характеристика приёмника).

AUC или area under curve — это просто площадь под кривой ROC. 

True Positive Rate (TPR) показывает, какой процент среди всех positive верно предсказан моделью.

$TPR = \frac{TP}{TP + FN}$

False Positive Rate (FPR): какой процент среди всех negative неверно предсказан моделью.

$FPR = \frac{FP}{TP + FN}$

```python
roc_auc_score(y_test, knn_y_pred)
```

### SVM

support vector machine, машина опорных векторов

Основная идея метода — перевод исходных векторов в пространство более высокой размерности и поиск разделяющей гиперплоскости с наибольшим зазором в этом пространстве. $P$ – число истинных результатов, $P = TP + FN$

$N$ – число ложных результатов, $N = TN + FP.$ 

* tp - количество истинных положительных результатов
* fn - количество ложных отрицательных результатов
* tn - количество истинных отрицательных результатов
* fp - количество ложных положительных результатов

```python
#accuracy = (количество верно классифицированных объектов) / (общее количество объектов)
#Оценка точности классификации

#The best value is 1 and the worst value is 0.
#precision = tp / (tp + fp)
#recall = tp / (tp + fn)
#F1 - среднее гармоническое между precision и recall
#F1 = 2 * (precision * recall) / (precision + recall)
```

```python
class Metrics:
    def accuracy(y_test, y_pred):
        TPN = 0 #TP+TN
        for i in range (len(y_test)): #len(y_test) - общее кол-во объектов
            #перебираем все объекты из тестовой выборки
            if y_test[i] == y_pred[i]: 
            #если объект из тестовой выборки равен объекту из предсказания
                TPN += 1 #количество верно классифицированных объектов +=1
        return (TPN / len(y_test))
    
    def precision(y_test, y_pred):
        TP, FP, FN, TN = 0, 0, 0, 0
        for i in range (len(y_test)):
          if y_test[i] == 0 and y_pred[i] == 1:
            FP += 1 #ложный положительный результат
          elif y_test[i] == 1 and y_pred[i] == 1:
            TP += 1 #истинный положительный
          elif y_test[i] == 1 and y_pred[i] == 0:
            FN += 1 #ложный отрицательный
          elif y_test[i] == 0 and y_pred[i] == 0:
            TN += 1 #истинный отрицательный
        return [[0, TN / (TN + FN)],
                [1, TP / (TP + FP)]]
    
    def recall(y_test, y_pred):
        TP, FP, FN, TN = 0, 0, 0, 0
        for i in range (len(y_test)):
          if y_test[i] == 0 and y_pred[i] == 1:
            FP += 1
          elif y_test[i] == 1 and y_pred[i] == 1:
            TP += 1
          elif y_test[i] == 1 and y_pred[i] == 0:
            FN += 1
          elif y_test[i] == 0 and y_pred[i] == 0:
            TN += 1
        return [[0, TN / (TN + FP)],
                [1, TP / (TP + FN)]]
    
    def F1_score(y_test, y_pred):
        B = 1
        prec = precision(y_test, y_pred)
        rec = recall(y_test, y_pred)
        return [[0, (1 + B ** 2) * prec[0][1] * rec[0][1] / ((B ** 2 * prec[0][1]) + rec[0][1])],
                [1, (1 + B ** 2) * prec[1][1] * rec[1][1] / ((B ** 2 * prec[1][1]) + rec[1][1])]]
```

### Реализация knn

```python
def euclideanDistance(x1, x2): #ограничим евклидово расстояние фиксированной длиной, игнорируя конечный размер
    return np.sqrt(sum((x1 - x2)**2))#определяем евклидово расстояние (расстояние между двумя точками в пространстве)

#Теперь, когда у нас есть мера сходства, мы можем использовать ее, чтобы собрать k наиболее похожих экземпляров для данного невидимого экземпляра.

#Это простой процесс вычисления расстояния для всех экземпляров и выбора подмножества с наименьшими значениями расстояния
```

```python
#Вычислим расстояния до всех точек обучающей выборки и отберем k соседей 
#(то есть тех, расстояния до которых минимальны)
def get_neighbors(x_train, x_test, k):
    distances = [(x_train[i][-1], euclidean_distance(x_train[i], x_test))
                  for i in range (len(x_train))]
    distances.sort(key=lambda elem: elem[1])
    
    neighbors = [distances[i][0] for i in range (k)]
    return neighbors
```

```python
#Теперь получим прогноз на основе классов соседей. 
#Посчитаем, сколько объектов каждого класса присутствует среди k ближайших к целевому, 
#и затем причислим его к тому классу, экземпляров которого больше всего.
def prediction(neighbors):
    count = {}
    for i in neighbors:
        if i in count:
            count[i] +=1
        else :
            count[i] = 1
    target = max(count.items(), key=lambda x: x[1])[0]
    return target
```

```python
#вычисляем среднее значение каждого атрибута для значения класса. 
#Среднее значение - это центральная средняя или центральная тенденция данных, 
#и мы будем использовать ее как середину нашего гауссовского распределения при расчете вероятностей.
def mean(numbers):
        return sum(numbers)/float(len(numbers))
    
#стандартное отклонение каждого атрибута для значения класса. 
#Стандартное отклонение описывает изменение разброса данных, 
#и мы будем использовать его для характеристики ожидаемого разброса каждого атрибута в нашем распределении Гаусса при расчете вероятностей.
#Стандартное отклонение рассчитывается как квадратный корень из дисперсии. 
#Дисперсия рассчитывается как среднее значение квадратов разностей для каждого значения атрибута из среднего значения
def stds(numbers):
        avg = mean(numbers)
        variance = sum([pow(x-avg,2) for x in numbers])/float(len(numbers)-1)
        return math.sqrt(variance)

#Теперь у нас есть инструменты для обобщения набора данных. 
#Для заданного списка экземпляров мы можем рассчитать среднее значение и стандартное отклонение для каждого атрибута    
def summa(data):
        summ = [(mean(attribute), stds(attribute)) for attribute in zip(*data)]
#zip группирует значения для каждого атрибута в наших экземплярах данных в свои собственные списки, чтобы мы могли вычислить среднее значение и значения стандартного отклонения для атрибута.
        del summ[-1] #удаляем
        return summ

#Мы можем объединить все это, сначала разделив наш обучающий набор данных на экземпляры, сгруппированные по классам. 
#Затем рассчитать итоги для каждого атрибута.
def summarizeByClass(data):
        separated = separateByClass(data)
        summ = {}
        for classValue, instances in separated.iteritems():
                summaries[classValue] = summarize(instances)
        return summ

#гауссовскую функцию для оценки вероятности данного значения атрибута, учитывая известное среднее значение и 
#стандартное отклонение для атрибута, оцененного по данным обучения.

def Probability(x, mean, stds): #сначала вычисляем показатель, затем вычисляем основное деление
        exponent = math.exp(-(math.pow(x-mean,2)/(2*math.pow(stds,2))))
        return (1 / (math.sqrt(2*math.pi) * stds)) * exponent


#объединяем вероятности всех значений атрибута для экземпляра данных и получаем вероятность того, что весь экземпляр данных принадлежит классу.

def ClassProbabilities(summaries, inputVector):
        probabilities = {}
        for classValue, classSummaries in summ.i():
            probabilities[classValue] = 1
            for i in range(len(classSummaries)):
                mean, stds = classSummaries[i]
                x = inputVector[i]
                probabilities[classValue] *= Probability(x, mean, stds)
        return probabilities
    

#делаем прогноз
def predict(summaries, inputVector):
        probabilities = ClassProbabilities(summ, inputVector)
        bestLabel, bestProb = None, -1
        for classValue, probability in probabilities.i():
            if bestLabel is None or probability > bestProb:
                bestProb = probability
                bestLabel = classValue
        return bestLabel
```
